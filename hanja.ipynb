```python
import hanja
import unicodedata, re, csv, os, sys
import pandas as pd
from tqdm import tqdm
```

코드 참고 : [github(hanja_hangul)](https://github.com/masoris/hanja_hangul)

```python
def only_hanja(str):
    '''한자만 추출'''
    str = unicodedata.normalize('NFC',str)
    len = '' 
    for i in str:
        uname = unicodedata.name(i)
        if uname[:3]=="CJK": len+=i
        if uname[:6]=="KANGXI": len+=i
    return len
```

```python
def len_hanja(str):
    '''한자 갯수 세기'''
    str = unicodedata.normalize('NFC',str)
    len = 0 
    for i in str:
        try:
            uname = unicodedata.name(i)
        except:
            print("문제 발생!")
            print(f'i : {i} / uname : {uname}')
        if uname[:3]=="CJK": len+=1
        if uname[:6]=="KANGXI": len+=1
    return len
```

### 한자 전처리

```python
train = pd.read_csv("/data/ephemeral/level2_datacentric/data/train_sp.csv")
train.head(2)
```

```python
cnt = 0
for row in train.iterrows():
    if len_hanja(row[1][1])>0 :
        cnt +=1 
        print(f'{cnt} : {row[1][1]}')
        print(f"{cnt} : {hanja.translate(row[1][1], 'substitution')}\n")
        

cnt
```

```python
def hanja2hangul(text):
    if len_hanja(text)>0:
        text = hanja.translate(text, 'substitution')
    return text
```

```python
tqdm.pandas()
train['text']=train['text'].progress_apply(hanja2hangul)
```

```python
train.to_csv("/data/ephemeral/level2_datacentric/data/train_sp_hanja.csv")
```
