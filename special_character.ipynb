```python
sp_dict = {}
for row in sp.iterrows():
    sp_dict[row[1][0]] = row[1][2]
sp_dict
```

```python
sp_dict['·'] = ' '
sp_dict['→'] = '에서 '
sp_dict['“'] = '"'
sp_dict['”'] = '" '
sp_dict['’'] = ' '
```

```python
def special_character(text):
    temp = []
    for k,v in sp_dict.items():
        if k in text:
            text = text.replace(k,v)
            temp.append(k)
    print(f'문장에 포함된 특수기호 : {temp}')
    print(f'문장 : {text}\n')
    return text
```

```python
df['text'] = df['text'].apply(special_character)
```

```python
def only_hanja(str):
    '''한자만 추출'''
    str = unicodedata.normalize('NFC',str)
    len = '' 
    for i in str:
        uname = unicodedata.name(i)
        if uname[:3]=="CJK": len+=i
        if uname[:6]=="KANGXI": len+=i
    return len
```

```python
def len_hanja(str):
    '''한자 갯수 세기'''
    str = unicodedata.normalize('NFC',str)
    len = 0 
    for i in str:
        try:
            uname = unicodedata.name(i)
        except:
            print("문제 발생!")
            print(f'i : {i} / uname : {uname}')
        if uname[:3]=="CJK": len+=1
        if uname[:6]=="KANGXI": len+=1
    return len
```

```python
only_hanja('대한민국(大韓民國)이라는 국호의 어원과 의미')
```

```python
hanja.translate('대한민국(大韓民國)이라는 국호의 어원과 의미', 'substitution')
```

```python
cnt = 0
for row in df.iterrows():
    if len_hanja(row[1][1])>0 :
        cnt +=1 
        print(f'{cnt} : {row[1][1]}')
        print(f"{cnt} : {hanja.translate(row[1][1], 'substitution')}\n")
        

cnt
```

```python
def hanja2hangul(text):
    if len_hanja(text)>0:
        text = hanja.translate(text, 'substitution')
    return text
```

```python
tqdm.pandas()
df['text']=df['text'].progress_apply(hanja2hangul)
```
